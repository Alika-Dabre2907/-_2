```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('nyc.csv')
```

```python
#removing duplicate
df=df.drop_duplicates()
df
```

```python
#deleting columns that would be irrelevant for analysis
df=df.drop(['name','neighbourhood','calculated_host_listings_count'],axis=1)
df.head()
```

```python
#removing duplicates
df=df.drop_duplicates()
df
```

```python
#checking for null values,dtype in all columns
df.info()
```

```python
#converting dtype and removing nan values
df['host_name']=df['host_name'].astype('string')
df['host_name']=df['host_name'].str.replace("nan"," ")

#ewmoving  non alphabetical character
df['host_name']=df['host_name'].str.replace(r'[^a-zA-Z\s]','',regex=True)
df

#creating space between 2 names
df['host_name'] = df['host_name'].str.replace(r'(\w)([A-Z])', r'\1 \2', regex=True)
df
```

```python
#obtaining data where Airbnb is priced
df=df[df['price']!=0]
df.head()
```

```python
#if we need data for minimum nights in a year or less
df=df[df['minimum_nights']<=365]
df
```

```python
#data where availability is atleast 1 day
df=df[df['availability_365']!=0]
df
```

```python
#converting data type to string
df['neighbourhood_group']=df['neighbourhood_group'].astype('string')
df['neighbourhood_group']

#deleting non values from
df['neighbourhood_group']=df['neighbourhood_group'].str.replace("nan"," ")
df
```

```python
#converting datatype to string
df['room_type']=df['room_type'].astype('string')
df['room_type']

#deleting non values
df['room_type']=df['room_type'].str.replace("nan"," ")
df
```
